```python
Xtr, Xtest, ytr, ytest = train_test_split(data, y, test_size=0.20, random_state=5)
```

```python
"""LOGISTIC REGRESSION"""
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import log_loss
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
from sklearn.metrics import accuracy_score
```

```python
clf=(LogisticRegression(C=0.02))
```

```python
clf.fit(Xtr, ytr)
```

```python
log_pred = clf.predict(Xtest)
```

```python
print(accuracy_score(clf.predict(Xtest), ytest))
```

```python
print(precision_score(ytest, log_pred, average = 'micro'))
```

```python
print(recall_score(ytest, log_pred, average = 'micro'))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest, log_pred)
```

```python
"""Random Forest"""
```

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
#Random Forest model
clfrf = RandomForestClassifier(max_features="log2", max_depth=11, n_estimators=24,min_samples_split=1000, 
                               oob_score=True)
```

```python

#fitting
clfrf.fit(Xtr, ytr)
```

```python
#predictions
predrf=clfrf.predict(Xtest)
```

```python
#accuracy score for the random forest model
accuracy_score(predrf, ytest)
```

```python
plt.figure(figsize=(50,8))
feature_imp_reordered = pd.Series(clfrf.feature_importances_,index= data.columns)
feature_imp_reordered.sort_values(ascending=False).plot(kind='Bar')
```

```python
features = pd.DataFrame()
features['feature'] = data.columns
features['importance'] = clfrf.feature_importances_
features.sort_values(by=['importance'], ascending=True, inplace=True)
features.set_index('feature', inplace=True)
```

```python
imp = features.sort_values(['importance'], ascending=False)
topf = 80
top50features = imp [:topf]
top50features.sort_values(by=['importance'], ascending=True, inplace=True)
top50features.plot(kind='barh', figsize=(25, 50))
```
